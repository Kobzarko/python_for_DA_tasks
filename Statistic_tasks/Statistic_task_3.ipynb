1 Скільки різних паролів можна скласти, використовуючи три літери англійського алфавіту, якщо літери не повинні повторюватися?

n=26 — загальна кількість літер англійського алфавіту.

r=3 — кількість літер у паролі.

$P(pass) = n!/(n-r)!= 26!/(26-23)!=26!/3!=26*25*24 = 15600$

2 Є група з 8 людей. Скількома способами можна обрати комітет із 3 осіб?

n=8 — загальна кількість людей у групі.

r=3 — кількість людей, яких потрібно вибрати для комітету.

$C(8,3)= 8!/3!(8-3)! = 8!/3!5! = 8*7*6*5*4*3*2*1/6*120 = 8*7*6/6 = 8*7=56$

3 Скількома різними способами можна розташувати на полиці 5 різних книг?

n = 5 кількість елементів, які потрібно розташувати

$C(5) = 5! =120$

4 В урні знаходяться 4 червоні, 3 сині і 2 зелені кулі. Скількома способами можна витягти 3 кулі, якщо вони повинні бути різних кольорів?

Щоб знайти загальну кількість способів витягти по одній кулі кожного кольору, треба перемножити кількість можливостей для кожного кольору:

$C (загальна​) =C (червоні​) * C(сині)​ * C (зелені​) =4* 3 * 2=24$
