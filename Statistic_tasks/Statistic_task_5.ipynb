### Домашнє завдання №5

1. У медичному центрі проводять тест на певну хворобу. Відомо, що 1% населення страждає на цю хворобу. Тест дає позитивний результат у 99% випадків для хворих і 5% випадків для здорових людей (хибнопозитивний результат). Якщо тест показав позитивний результат, яка ймовірність того, що людина насправді хвора?

Дані:

P(H)=0.01 — ймовірність того, що людина хвора (1% населення).

P(¬H)=0.99 — ймовірність того, що людина здорова.

P(T+∣H)=0.99 — ймовірність того, що тест позитивний для хворих (99%).

P(T+∣¬H)=0.05 — ймовірність того, що тест позитивний для здорових (5% хибнопозитивних результатів).

Потрібно знайти: 

P(H∣T+) — ймовірність того, що людина хвора за умови, що тест позитивний.

Теорема Баєса:

$P(H|T+) = P(T+|H) * P(H)/ P(T+)$

Де: 

P(T+) — це загальна ймовірність позитивного тесту, яку можна обчислити як:

$P(T+)=P(T+∣H)*P(H)+P(T+∣¬H)*P(¬H)$

Обчислимо:

$P(T+)=(0.99⋅0.01)+(0.05⋅0.99) = 0.0099 + 0.0495 = 0.0594$


$P(H|T+) = 0.99 * 0.01/0.0594 = 0.1667$

Висновок:

Якщо тест дав позитивний результат, ймовірність того, що людина насправді хвора, дорівнює приблизно 16.67%.

2. У компанії 70% співробітників чесні, а 30% можуть інколи брехати. Детектор брехні правильно ідентифікує чесних співробітників у 90% випадків і неправильно класифікує їх як таких, що брехали, у 10% випадків. Якщо детектор показав, що співробітник брехав, яка ймовірність того, що він насправді брехав?

Дані:

P(H)=0.7 — ймовірність того, що співробітник чесний (70%).

P(B)=0.3 — ймовірність того, що співробітник брехливий (30%).

P(T+∣H)=0.1 — ймовірність того, що чесний співробітник буде помилково визнаний брехливим (10% помилкових позитивних результатів).

P(T+∣B)=1 — ймовірність того, що брехливий співробітник буде правильно визнаний брехливим (100%).

Потрібно знайти: P(B∣T +) — ймовірність того, що співробітник насправді брехав, якщо детектор показав, що він брехав.

Теорема Баєса:

$P(B|T+) = P(T+|B) * P(B)/ P(T+)$

Де P(T+) - це загальна ймовірність позитивного тесту, яку можна обчислити як:

$P(T+)=P(T+∣H)*P(H)+P(T+∣B)*P(B)$

$P(T+) = 0.1 * 0.7 + 1 * 0.3 = 0.37$

$P(B|T+) = 1 * 0.3 / 0.37 = 0.3/0.37 = 0.81$

Висновок:

Ймовірність того, що співробітник насправді брехав, якщо детектор показав, що він брехав, дорівнює приблизно 81.08%.

3. Відомо, що предмет можна загубити в одній з трьох кімнат: у першій кімнаті з ймовірністю 0,5, у другій — з ймовірністю 0,3, і в третій — з ймовірністю 0,2. Якщо предмет загублено в кімнаті, ймовірність знайти його під час пошуку складає 0,8 для першої кімнати, 0,6 для другої і 0,4 для третьої. Якщо предмет знайшли, яка ймовірність того, що його було загублено в першій кімнаті?

Дані:

Ймовірності загублення предмета в кожній кімнаті:

P(A1)=0.5 — ймовірність загублення в першій кімнаті.

P(A2)=0.3 — ймовірність загублення в другій кімнаті.

P(A3)=0.2 — ймовірність загублення в третій кімнаті.

Ймовірності знайдення предмета в кожній кімнаті:

P(F|A1)=0.8 — ймовірність знайти предмет у першій кімнаті, якщо його там загубили.

P(F|A2)=0.6 — ймовірність знайти предмет у другій кімнаті, якщо його там загубили.

P(F|A3)=0.4 — ймовірність знайти предмет у третій кімнаті, якщо його там загубили.

Потрібно знайти: 

P(A1∣F) — ймовірність того, що предмет було загублено в першій кімнаті, якщо його знайшли.

Теорема Баєса:

$P(A1∣F) = P(F|A1) * P(A1)/ P(F)$

Де P(F) - це загальна ймовірність позитивного тесту, яку можна обчислити як:

$P(F)=P(F|A1)*P(A1)+P(F|A2)*P(A2)+P(F|A3)*P(A3)$

$P(F)= 0.8 * 0.5 + 0.6 * 0.3 + 0.4 * 0.2 = 0.4 + 0.18 + 0.08 = 0.66$

$P(A1∣F) = 0.8 * 0.5 / 0.66 = 0.4 / 0.66 = 0.6061$

Висновок:

Ймовірність того, що предмет було загублено в першій кімнаті, якщо його знайшли, дорівнює приблизно 60.61%.

4. На заводі виробляють продукцію на трьох лініях: 50% продукції випускає перша лінія, 30% — друга, і 20% — третя. Відомо, що 2% продукції з першої лінії, 3% з другої і 5% з третьої мають дефекти. Якщо випадково обрана продукція має дефект, яка ймовірність того, що вона була випущена третьою лінією?

Дані:

Ймовірності випуску продукції з кожної лінії:

`P(L1​)=0.5` — ймовірність, що продукція випущена з першої лінії.

`P(L2​)=0.3` — ймовірність, що продукція випущена з другої лінії.

`P(L3​)=0.2` — ймовірність, що продукція випущена з третьої лінії.

Ймовірності дефектної продукції з кожної лінії:

`P(D∣L1​)=0.02` — ймовірність дефекту для продукції з першої лінії.

`P(D∣L2​)=0.03` — ймовірність дефекту для продукції з другої лінії.

`P(D∣L3​)=0.05` — ймовірність дефекту для продукції з третьої лінії.

Потрібно знайти: 

`P(L3​∣D)` — ймовірність того, що продукція була випущена третьою лінією, якщо вона має дефект.

Теорема Баєса:

$P(L3​∣D) = P(D∣L3) * P(L3)/ P(D)$

Де `P(D)` - це загальна ймовірність позитивного тесту, яку можна обчислити як:

$P(D)=P(D∣L1)*P(L1)+P(D∣L2)*P(L2)+P(D∣L3)*P(L3)$

$P(D) = 0.02*0.5 + 0.03*0.3 + 0.05*0.2 = 0.01 + 0.009 + 0.01 = 0.029$

Тепер підставимо значення в теорему Баєса:

$P(L3|D) = 0.05 * 0.2 / 0.029 = 0.01 / 0.029 = 0.3448$

Висновок:

Ймовірність того, що продукція з дефектом була випущена третьою лінією, дорівнює приблизно 34.48%.


5. Програма для розпізнавання рукописних цифр має точність 95% для цифри "3" і 98% для інших цифр. Якщо на вхід програми подано цифру "3", а програма її не розпізнала, яка ймовірність того, що насправді була подана цифра "3"? Відомо, що в 20% випадків на вхід подається цифра "3".

Дані:

Ймовірності випуску продукції з кожної лінії:

`P(3​)=0.2` — ймовірність того, що на вхід програми подається цифра "3".

`P(¬3​)=1 - P(3​) = 0.8` — ймовірність того, що на вхід програми подається не цифра "3".

`P(¬R|3​)=1 - 0.95 = 0.05` — ймовірність того, що програма не розпізнає цифру "3", якщо насправді подано "3".

`P(¬R|¬3​)=1 - 0.98 = 0.02` — ймовірність того, що програма не розпізнає цифру, якщо насправді подано не "3". Якщо програма має точність 98% для інших цифр, то ймовірність не розпізнати (помилкове невизнання) цифри, яка не є "3".

Знайти:

Необхідно знайти ймовірність того, що насправді була подана цифра "3", якщо програма не розпізнала:

$P(3|¬R​)$

Теорема Баєса:

$P(3|¬R​) = P(¬R|3​) * P(3​)/ P(¬R)$

Де `P(¬R)` - це загальна ймовірність позитивного тесту, яку можна обчислити як:

$P(¬R)=P(¬R|3​)*P(3) + P(¬R|¬3​)*P(¬3​)$

$P(¬R)= 0.05*0.2 + 0.02*0.8 = 0.01 + 0.016 = 0.026$

Тепер підставимо значення в теорему Баєса:

$P(3|¬R​) = 0.05 * 0.2 / 0.026 = 0.01/0.026 = 0.3846$

Висновок:

Ймовірність того, що насправді була подана цифра "3", якщо програма її не розпізнала, дорівнює приблизно 38.46%.
