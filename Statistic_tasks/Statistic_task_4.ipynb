### Домашнє завдання №4

1. З колоди у 52 карти випадково витягується одна карта. Яка ймовірність того, що це туз, якщо відомо, що ця карта — пікова?

Подія A: карта — туз.

Подія B: карта — пікова

В колоді 52 карти, з яких 13 — пікові. Тому ймовірність того, що витягнута карта — пікова:

$P(B) = 13/52 = 1/4$

Ймовірність одночасного настання подій A∩B (пікова карта і туз): В колоді є тільки один піковий туз. Отже, ймовірність того, що карта є одночасно піковою і тузом:

$P(A∩B)= 1/52$

Знайдемо умовну ймовірність P(A∣B):

$P(A∣B) = P(A∩B) / P(B) = 1/52 / 13/52 = 1/13$



2. У групі з 30 студентів є 18 хлопців і 12 дівчат. Відомо, що 7 хлопців і 5 дівчат носять окуляри. Випадково обраний студент носить окуляри. Яка ймовірність того, що цей студент — дівчина?

Подія A: студент — дівчина.

Подія B: студент носить окуляри. 

У групі з 30 студентів, 5+7 носять окуляри:

$P(B) = 12/30 = 2/5$

Кількість дівчат, що носять окуляри, — 5. Тому ймовірність того, що випадково обраний студент є одночасно дівчиною і носить окуляри:

$P(A∩B) = 5/30 = 1/6$

Знайдемо умовну ймовірність P(A∣B):

$P(A∣B) = P(A∩B) / P(B) = 1/6 / 2/5 = 5/12$

3. У конверті знаходяться 10 лотерейних білетів, серед яких 3 виграшних. З конверта послідовно витягуються білети. Знайти ймовірності того, що:
    
    а) 2-й витягнутий білет буде виграшним, якщо 1-й був виграшним;
    б) 3-й буде виграшним, якщо попередні два білети були виграшними;
    в) 4-й буде виграшним, якщо попередні білети були виграшними.

Є 10 лотерейних білетів.

Серед них 3 виграшних і 7 невиграшних.

Білети витягуються послідовно, без повернення.

а) Ймовірність того, що 2-й витягнутий білет буде виграшним, якщо 1-й був виграшним.

Якщо 1-й білет був виграшним, то в конверті залишаються 9 білетів, з яких 2 — виграшних.

Тоді ймовірність того, що 2-й витягнутий білет буде виграшним:

$P(A|B) = 2/9$

б) Ймовірність того, що 3-й витягнутий білет буде виграшним, якщо попередні два були виграшними.

Якщо 2 попередні білети були виграшними, то в конверті залишається 8 білетів, з яких 1 — виграшний.

Ймовірність того, що 3-й витягнутий білет буде виграшним:

$P(A|B) = 1/8$

в) Ймовірність того, що 4-й витягнутий білет буде виграшним, якщо попередні три були виграшними.

Якщо 3 попередні білети були виграшними, то в конверті залишаються 7 білетів, і всі вони невиграшні.

Тому ймовірність того, що 4-й витягнутий білет буде виграшним дорівнює 0.

$P(A|B) = 0$
