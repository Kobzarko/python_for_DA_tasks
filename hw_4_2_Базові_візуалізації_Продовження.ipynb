### Завдання: Візуалізація функцій експоненти, логарифму, і модифікованої синусоїди

**Мета:** Створити фігуру з трьома панелями: на першій показати графік експоненціальної функції, на другій - логарифмічну функцію, а на третій - модифіковану синусоїду.

1. Імпортуйте необхідні бібліотеки:


```python
import numpy as np
import matplotlib.pyplot as plt
```


2. Використовуючи фукнціонал `numpy` створіть масив `x` з 400 точок в діапазоні між -2 та 2 і масив `x_positive` в діапазоні між 0.1 і 2 також з 400 точок. Другий масив в нас буде для логарифмічної функції, перший з більш широким діапазоном - для інших функцій.


```python
x = np.linspace(-2, 2, 400)
x_positive = np.linspace(0.1, 2, 400)
```


3. Тут маємо кілька кроків.

  3.1. Ініціалізуйте фігуру Matplotlib з розміром, достатнім для трьох панелей (рекомендую спробувати експериментально виявити добрий розмір).   
  3.2. На першій панелі візуалізуйте експоненціальну функцію `y = e^x`. Використовуйте відповідну змінну `x`.  
  3.2. На другій панелі покажіть логарифмічну функцію `y = ln(x)`.  
  3.4. На третій панелі покажіть модифіковану синусоїду `y = sin(x) + x`.  

  Відобразіть графік.


```python
plt.figure(figsize=(14, 4))  # створюємо фігуру

# створюємо першу з двох панелей і встановлюємо поточну вісь
plt.subplot(1, 3, 1)  # (рядки, колонки, номер панелі)
plt.plot(x, np.exp(x))

# створюємо другу з двох панелей і встановлюємо поточну вісь
plt.subplot(1, 3, 2);
plt.plot(x, np.log(x_positive));

# створюємо третю з двох панелей і встановлюємо поточну вісь
plt.subplot(1, 3, 3);
plt.plot(x, np.sin(x) + x);

plt.tight_layout();
```

4. Модифікація і довершення графіку.  
  4.1. Для кожної панелі встановіть назву графіка та підписи до осей х та у (можна так і назвати `x` та формула функції).  
  4.2. Використайте різні кольори для кожного графіка.  


```python
# створюємо першу з трьох панелей
plt.figure(figsize=(12, 3))
plt.subplot(1, 3, 1)  # (рядки, колонки, номер панелі)
plt.plot(x, np.exp(x),'-g', label='y = e^x')
plt.title('Експоненціальна функція')
plt.xlabel('Ось X', fontsize=10, color='green')
plt.ylabel('Ось Y', fontsize=10, color='green')
plt.tight_layout();
```

```python
# створюємо другу з трьох панелей
plt.figure(figsize=(12, 3))
plt.subplot(1, 3, 2);
plt.plot(x, np.log(x_positive),'-b', label='y = ln(x)');
plt.title('Логарифмічна функція')
plt.xlabel('Ось X', fontsize=10, color='blue')
plt.ylabel('Ось Y', fontsize=10, color='blue')
plt.tight_layout();
```

```python
# створюємо третю з трьох панелей
plt.figure(figsize=(12, 3))
plt.subplot(1, 3, 3);
plt.plot(x, np.sin(x) + x,'-r', label='y = sin(x) + x');
plt.title('Модифікована синусоїда')
plt.xlabel('Ось X', fontsize=10, color='red')
plt.ylabel('Ось Y', fontsize=10, color='red')
plt.tight_layout();
```

5. А тепер зобразіть ті самі 3 трафіки тих самих кольорів тільки на одному графіку. Визначіть назву кожної з ліній для легенди та обовʼязково відобразіть легенду.

```python
plt.figure(figsize=(14, 4))  # створюємо фігуру

# створюємо першу з двох панелей і встановлюємо поточну вісь
plt.subplot(1, 3, 1)  # (рядки, колонки, номер панелі)
plt.plot(x, np.exp(x),'-g', label='y = e^x')
plt.title('Експоненціальна функція')
plt.xlabel('Ось X', fontsize=10, color='green', loc='right')
plt.ylabel('Ось Y', fontsize=10, color='green', loc='top')
plt.legend();

# створюємо другу з двох панелей і встановлюємо поточну вісь
plt.subplot(1, 3, 2);
plt.plot(x, np.log(x_positive),'-b', label='y = ln(x)')
plt.title('Логарифмічна функція')
plt.xlabel('Ось X', fontsize=10, color='blue', loc='right')
plt.ylabel('Ось Y', fontsize=10, color='blue', loc='top')
plt.legend();

# створюємо третю з двох панелей і встановлюємо поточну вісь
plt.subplot(1, 3, 3);
plt.plot(x, (np.sin(x) + x), '-r', label='y = sin(x) + x')
plt.title('Модифікована синусоїда')
plt.xlabel('Ось X', fontsize=10, color='red', loc='right')
plt.ylabel('Ось Y', fontsize=10, color='red', loc='top')
plt.legend();

plt.tight_layout();
```
