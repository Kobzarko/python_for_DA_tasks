В який день тижня Total найвищий

Буде зрозуміло в який день потрібно проводити інкасацію супермаркету або грабувати)
Також потрібно порівняти чи співпадає цей день з днем акції.

```python

# Перетворення стовпця "Date" на datetime тип
df['Date'] = pd.to_datetime(df['Date'])

day_name_map = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}

# Додавання стовпця з днем тижня
df['Day of Week'] = df['Date'].dt.dayofweek

df['Day of Week'] = df['Day of Week'].map(day_name_map)

# Створення зведеної таблиці
pivot_table_sales_by_day = df.pivot_table(values='Total', index='Product line',columns='Day of Week', aggfunc='sum')

max_category = pivot_table_sales_by_day.idxmax(axis=1)
max_value_category = pivot_table_sales_by_day.max(axis=1)

max_total_table = pd.DataFrame({
    'Day of Week': max_category,
    'Max Sales (Total)': max_value_category
})

# pivot_table_sales_by_day
max_total_table
```

Яка категорія товарів найпопулярніша в різних містах, а яка менш популярна. Вичислити за кількістю проданих товарів у кожнії категорії.

Цей аналіз дає глобальне розуміння яка категорія потребує уваги для поглибленого аналізу товарів.


```python

category_popularity_by_city = df.pivot_table(values='Quantity', index='City', columns='Product line', aggfunc='sum')

most_popular_category_by_city = category_popularity_by_city.idxmax(axis=1)
less_popular_category_by_city = category_popularity_by_city.idxmin(axis=1)
category_popularity_by_city['Most Popular Category'] = most_popular_category_by_city
category_popularity_by_city['Less Popular Category'] = less_popular_category_by_city
category_popularity_by_city

```

Яким методом оплати (Payment) користуються найбільше в різних категоріях.

Цей аналіз дає розуміння в яких категоріях можна використовувати кешбек при оплаті карткою.

```python
payment_method_table = df.pivot_table(values='Invoice ID', index='Product line', columns='Payment', aggfunc='count')

payment_method_table['Most Popular payment'] = payment_method_table.idxmax(axis=1)

payment_method_table
```
